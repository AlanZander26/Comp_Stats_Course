## 3. Exercise: Central Limit Theorem
Plot the distribution of the MLEs for `n = [1, 2, 5, 10, 100, 1000]` observations in each experiment, assuming the following distributions. The number `N` of experiments is up to you. Then fit a Gaussian to verify the agreement with the CLT.

### 3.1 Bernoulli: p = 0.5

```python
n_sample = [1, 2, 5, 10, 100, 1000]  # Different sample sizes
N = 10_000  # Number of experiments
p = 0.5 # parameter of Bernoulli distribution

# Create a 2x3 grid of subplots
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.ravel()  # Flatten axes array for easy indexing

# Loop over sample sizes and plot on each subplot
for i, n in enumerate(n_sample):
    # Generate exponential random variables and calculate means along axis 0
    bernoulli_means = np.random.binomial(1, p=p, size=(n, N)).mean(axis=0)
    
    # Plot histogram of sample means with only outline
    bins = 50 + 3*n**2//N
    axes[i].hist(bernoulli_means, bins=np.linspace(0, 1, bins), histtype='step', label=f'n={n}', density=True)
    
    # Calculate and plot corresponding normal distribution
    x = np.linspace(0, 1, 300)
    y = sp.stats.norm.pdf(x, loc=p, scale=np.sqrt(p*(1-p)/n))
    axes[i].plot(x, y, linestyle='-', linewidth=2, label=f'Normal Dist (n={n})')
    
    # Set limits, labels, and legend for each subplot
    axes[i].set_xlim(-0.01, 1.01)
    axes[i].set_title(f'Sample Size n={n}')
    axes[i].set_xlabel('Mean')
    axes[i].set_ylabel('Frequency')
    axes[i].legend()
    axes[i].grid()

# Adjust layout to prevent overlapping
plt.tight_layout()
plt.suptitle('Central Limit Theorem Demonstration - Bernoulli', y=1.05, fontsize=16)
plt.show()
```

### 3.2 Exponential: $\tau = 1$


```python
n_sample = [1, 2, 5, 10, 100, 1000]  # Different sample sizes
N = 10_000  # Number of experiments 
Lambda = 1

# Create a 2x3 grid of subplots
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.ravel()  # Flatten axes array for easy indexing

# Loop over sample sizes and plot on each subplot
for i, n in enumerate(n_sample):
    # Generate exponential random variables and calculate means along axis 0
    exp_means = np.random.exponential(scale=Lambda, size=(n, N)).mean(axis=0)
    
    # Plot histogram of sample means with only outline
    bins = 50 + 3*n**2//N
    axes[i].hist(exp_means, bins=np.linspace(0, 3, bins), histtype='step', label=f'n={n}', density=True)
    
    # Calculate and plot corresponding normal distribution
    x = np.linspace(0, 3, 300)
    y = sp.stats.norm.pdf(x, loc=1/Lambda, scale=1/np.sqrt(Lambda**2*n))
    axes[i].plot(x, y, linestyle='-', linewidth=2, label=f'Normal Dist (n={n})')
    
    # Set limits, labels, and legend for each subplot
    axes[i].set_xlim(0, 3)
    axes[i].set_title(f'Sample Size n={n}')
    axes[i].set_xlabel('Mean')
    axes[i].set_ylabel('Frequency')
    axes[i].legend()
    axes[i].grid()

# Adjust layout to prevent overlapping
plt.tight_layout()
plt.suptitle('Central Limit Theorem Demonstration - Exponential', y=1.05, fontsize=16)
plt.show()
```

## 4. Exercise: MC Integration

```python
# MC integration in 1 dim
def MC_integration_1D(f, a, b, n):
    V = b - a
    X = np.random.uniform(a, b, n)
    if not isinstance(f, np.ufunc):
        f = np.vectorize(f)
    F = f(X)
    I = V * F.mean()
    return I

# General MC integration in multiple dimensions
def MC_integration(f, a, b, n):
    a = np.array(a)
    b = np.array(b)
    V = np.prod(b - a)
    X = np.random.uniform(low=a, high=b, size=(n, a.size))
    if not isinstance(f, np.ufunc):
        f = np.vectorize(f)
    F = f(*X.T)
    I = V * F.mean()
    return I
```

## 4.1 Estimate the integral $I$ with $f(x) = e^x$ over the interval $[a, b] = [-1, 1]$.

```python
# Using the MC integration for 1 dim
f = np.exp
a = -1
b = 1
n = 10_000_000
MC_integration_1D(f, a, b, n)
```

```python
# Using general MC integration
f = np.exp
a = -1
b = 1
n = 10_000_000
MC_integration(f, a, b, n)
```

##  4.2 Plot the distribution of $\hat{I}$ for $N = 10, 100, 1000$.

```python
f = np.exp#lambda x : np.exp(x)*np.log(x)
a = -1
b = 1
I = np.exp(b) - np.exp(a)
n_list = [10, 100, 1000]
I_results = {}
for n in n_list:
    I_list = []
    for _ in range(1000):
        I = MC_integration(f, a, b, n)
        I_list.append(I)
    I_results[n] = np.array(I_list)

# Plot histograms on the same figure
plt.figure(figsize=(10, 6))
colors = ['blue', 'orange', 'green']
bins = 30

for idx, n in enumerate(n_list):
    plt.hist(I_results[n], bins=bins, color=colors[idx], alpha=0.5, label=f'n={n}', density=True)
    # Mean and standard deviation
    mean_I = I_results[n].mean()
    std_I = I_results[n].std()
    print(n, mean_I, std_I**2*n) # Multiply the variance by n to have an estimate of the variance of w
    x = np.linspace(mean_I - 4*std_I, mean_I + 4*std_I, 200)
    y = sp.stats.norm.pdf(x, mean_I, std_I)

    # Plot the normal distribution
    plt.plot(x, y, color=colors[idx], linestyle='-', linewidth=2, label=f'Normal Dist (n={n})')

    
plt.axvline(I, color='k', linestyle='--', label='True value')
plt.title('Monte Carlo Integration Results Comparison')
plt.xlabel('Integration Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid()
plt.show()
```
